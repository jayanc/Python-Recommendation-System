```python
# Import libraries
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import DataSet
df = pd.read_excel('./DataSets/online-retail-data-set-from-uci-ml-repo/Online Retail.xlsx', header = 0,nrows=1000)
```

```python
print('dataset dimensions are:', df.shape)
df.describe(include = 'all')
```

```python
cleaned_retail = df.loc[pd.isnull(df.CustomerID) == False]
cleaned_retail
```

```python
item_lookup = cleaned_retail[['CustomerID','StockCode','Quantity','InvoiceDate']].drop_duplicates()

item_lookup['StockCode'] = pd.to_numeric(item_lookup['StockCode'],errors='coerce').fillna(0).astype('int64')
item_lookup['CustomerID'] = item_lookup.CustomerID.astype('int64')
```

```python
item_lookup = item_lookup[(item_lookup[['StockCode']] != 0).all(axis=1)]

item_lookup['ratings'] = (item_lookup['Quantity'] / item_lookup['Quantity'])


item_lookup
```

```python
item_lookup.dtypes
```

```python
item_lookup
max_CustomerID = item_lookup['CustomerID'].drop_duplicates().max()+1
max_StockCode = item_lookup['StockCode'].drop_duplicates().max()+1
max_Quantity = item_lookup['Quantity'].drop_duplicates().max()+1
print('CustomerID:',max_CustomerID,'StockCode:',max_StockCode,'max_Quantity:',max_Quantity)
```

```python
# Create training set
shuffled_item_lookup = item_lookup.sample(frac=1.)
# print(shuffled_item_lookup)

# Shuffling CustomerID
new_CustomerID = shuffled_item_lookup['CustomerID'].values
# max_CustomerID= shuffled_item_lookup['CustomerID'].drop_duplicates().max()
print ('CustomerID:', new_CustomerID, ', shape =', new_CustomerID.shape,"max_CustomerID",max_CustomerID)

# Shuffling StockCode
new_StockCode = shuffled_item_lookup['StockCode'].values
# max_StockCode = shuffled_item_lookup['StockCode'].drop_duplicates().max()
print ('StockCode:', new_StockCode , ', shape =', new_StockCode.shape,'max_StockCode',max_StockCode)

# Shuffling Quantity
new_Quantity = shuffled_item_lookup['Quantity'].values
# max_Quantity = shuffled_item_lookup['Quantity'].drop_duplicates().max()
print ('Quantity:', new_Quantity , ', shape =', new_Quantity.shape,'max_Quantity',max_Quantity)
```

```python
# Import Keras libraries
from keras.callbacks import Callback, EarlyStopping, ModelCheckpoint
# Import CF Model Architecture
from CFModel import CFModel
```

```python
# Define constants
K_FACTORS = 100 # The number of dimensional embeddings for movies and users
TEST_USER = 13047 # A random test user (user_id = 2000)
```

```python
# Define model
model = CFModel(max_CustomerID,max_StockCode, K_FACTORS)
# Compile the model using MSE as the loss function and the AdaMax learning algorithm
model.summary()
model.compile(loss='mse', optimizer='adamax',metrics=['accuracy'])
```

```python
# Callbacks monitor the validation loss
# Save the model weights each time the validation loss has improved
callbacks = [EarlyStopping('val_loss', patience=2), 
             ModelCheckpoint('retail_weights.h5', save_best_only=True)]

# Use 30 epochs, 90% training data, 10% validation data 
history = model.fit([new_CustomerID, new_StockCode],new_Quantity,nb_epoch=30, validation_split=.1, verbose=2, callbacks=callbacks)
```

```python
# Show the best validation RMSE
min_val_loss, idx = min((val, idx) for (idx, val) in enumerate(history.history['val_loss']))
print('Minimum RMSE at epoch', '{:d}'.format(idx+1), '=', '{:.4f}'.format(math.sqrt(min_val_loss)))
```

```python
# Use the pre-trained model
trained_model = CFModel(max_CustomerID,max_StockCode, K_FACTORS)
# Load weights
trained_model.load_weights('retail_weights.h5')
```

```python
# Pick a random test user
item_lookup[item_lookup['CustomerID'] == TEST_USER]
```

```python
# Function to predict the ratings given User ID and Movie ID
def predict_rating(user_id, movie_id):
    return trained_model.rate(user_id - 1, movie_id - 1)
```

```python
user_ratings = item_lookup[item_lookup['CustomerID'] == TEST_USER][['CustomerID','StockCode','Quantity']]
user_ratings['prediction'] = user_ratings.apply(lambda x: predict_rating(TEST_USER, x['StockCode']), axis=1)
user_ratings.sort_values(by='Quantity', 
                         ascending=False).head(20)
```

```python
recommendations = item_lookup[item_lookup['CustomerID'].isin(user_ratings['CustomerID']) == False][['StockCode']].drop_duplicates()
recommendations['prediction'] = recommendations.apply(lambda x: predict_rating(TEST_USER, x['StockCode']), axis=1)
recommendations.sort_values(by='prediction',ascending=False)
# .merge(movies,
#                                                  on='movie_id',
#                                                  how='inner',
#                                                  suffixes=['_u', '_m']).head(20)
```
